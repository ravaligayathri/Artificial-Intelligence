```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from collections import Counter
import numpy as np
import operator
from collections import defaultdict

class DecisionStump:
    
    def entropy(self,entro):
        entropy = 0
        val, counts = np.unique(entro, return_counts=True)
        freqs = counts.astype('float')/len(entro)
        for prob in freqs:
            if prob != 0:
                entropy = entropy - prob * np.log2(prob)
        
        return entropy  
    
    def info_gain(self,y, x):
        gain = self.entropy(y)
        values, counts = np.unique(x, return_counts=True)
        freqs = counts.astype('float')/len(x)
        for freq, value in zip(freqs, values):
            gain = gain- freq * self.entropy(y[x == value])
        
        return gain
    
    
    def partition(self,a):
            
        return {c: (a==c).nonzero()[0] for c in np.unique(a)}
    
    def fit(self,x,y):
        self.Y = Y
        if(len(set(Y)) == 1):
            return y
        gain = np.array([self.info_gain(Y,val) for val in X.T])
        self.feature = np.argmax(gain)
        self.sets = self.partition(X[:, self.feature])
    ct = defaultdict(list)
    
    
    def predict(self,testdata):
        result = []
        a = testdata[:,self.feature]
        for i in range(0,len(a)):
            val = self.sets.get(a[i])
            sub = self.Y.take(val, axis=0)
            target = Counter(sub)
            result.append(max(target.items(), key=operator.itemgetter(1))[0])
        #print(result)
        return result


X = np.array([['sunny','hot','high','weak'],
        ['sunny','hot','high','strong'],
        ['overcast','hot','high','weak'],
        ['rain','mild','high','weak'],
        ['rain','cool','normal','weak'],
        ['rain','cool','normal','strong'],
        ['overcast','cool','normal','strong'],
        ['sunny','mild','high','weak'],
        ['sunny','cool','normal','weak'],
        ['rain','mild','normal','weak'],
        ['sunny','mild','normal','strong'],
        ['overcast','mild','high','strong'],
        ['overcast','hot','normal','weak'],
        ['rain','mild','high','strong'],
        ['sunny','mild','high','strong']
    ])

Y = np.array(['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no','no'])

    
    
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

print(X_train)
print(y_train)


Decisionstump = DecisionStump()

Decisionstump.fit(X_train,y_train)

y_pred = Decisionstump.predict(X_test)

print('Accuracy:',accuracy_score(y_test, y_pred)*100)

```
